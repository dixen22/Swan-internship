```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("data/Admission_Predict.csv", index_col="Serial No.")
df
```

```python
sns.pairplot(df, y_vars="Chance of Admit")
```

```python
df_train, df_test = train_test_split(df[["Chance of Admit", "GRE Score", "TOEFL Score", "CGPA"]], train_size=0.8)
df_valid, df_eval = train_test_split(df_test, train_size=0.5)
df_train.to_csv("data/train.csv")
df_valid.to_csv("data/valid.csv")
df_eval.to_csv("data/eval.csv")
```
