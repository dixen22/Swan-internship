28. Consider a random 10x2 matrix representing cartesian coordinates, convert them to polar coordinates (★★☆)

```haskell
coord <- randintIO' 0 10 [10, 2]
coord
```

```haskell
cartToPol :: Tensor -> Tensor
cartToPol t =
    let x = Torch.select 1 0 t
        y = Torch.select 1 1 t
        r = Torch.sqrt (x * x + y * y) 
        theta = Torch.Functional.Internal.atan2 y x
    in Torch.stack (Dim 1) [r, theta]

cartToPol coord
```

34. How to find the closest value (to a given scalar) in a vector? (★★☆)

```haskell
v <- randIO' [10]
v
```

```haskell
getClosestVal :: Float -> Tensor -> Tensor
getClosestVal n t = 
    let target = asTensor (n :: Float)
        dist = Torch.abs (t - target)
        closest_i = argmin dist 0 False
    in Torch.select 0 (asValue closest_i :: Int) t
    
getClosestVal 0.5 v
```

42. How to sort an array by the nth column? (★★☆)

```haskell
m <- randIO' [5, 5]
m
```

```haskell
sortByCol :: Int -> Tensor -> Tensor
sortByCol n t =
    let col = Torch.select 1 n t
        sorted_i = snd (sort col 0 False)
    in Torch.indexSelect 0 sorted_i t
    
sortByCol 2 m
```
